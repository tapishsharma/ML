```python
#libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Importing dataset
#customer id column is useless here so we can remove it.

dataset=pd.read_csv("F:\ML LEARNINGS\Clustering\Mall_Customers.csv")
x=dataset.iloc[:,[3,4]].values
```

```python
# Using elbow method to find the optimal number of clusters
from sklearn.cluster import KMeans
wcss = []

# here we will appending the wcss(Within-Cluster-Sum-of-Squares) .sum of squares of the distances of each data point in all clusters to their respective centroids
for i in range(1,11):
    kmeans = KMeans(n_clusters=i,init = 'k-means++', random_state=42)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)

    #plotting elbow graph

plt.plot(range(1,11),wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('wcss')
plt.show()
```

```python
#Traiing the k-means model

kmeans = KMeans(n_clusters=5,init = 'k-means++', random_state=42)
#creating dependent variable also while fitting
y_dependent = kmeans.fit_predict(x)

```

```python
print(y_dependent)
# So here 5 clusters ar created 0==1st cluster and so on ,,, all customers are divided upon clusters
```

```python
#Plotting

plt.scatter(x[y_dependent== 0, 0], x[y_dependent == 0, 1], s = 50, c = 'red', label = 'Cluster 1')
plt.scatter(x[y_dependent == 1, 0], x[y_dependent == 1, 1], s = 50, c = 'blue', label = 'Cluster 2')
plt.scatter(x[y_dependent == 2, 0], x[y_dependent == 2, 1], s = 50, c = 'green', label = 'Cluster 3')
plt.scatter(x[y_dependent == 3, 0], x[y_dependent == 3, 1], s = 50, c = 'cyan', label = 'Cluster 4')
plt.scatter(x[y_dependent == 4, 0], x[y_dependent == 4, 1], s = 50, c = 'magenta', label = 'Cluster 5')

#plotting centroids now 
#cluster_centers stores a 2d matrix of centroids with coordinates
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 200, c = 'yellow', label = 'Centroids')
plt.title('Clusters of customers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```
